# @classmethod and @staticmethod

# classmethod

```python
class PlayerBall:
    # Class object attribute
    membership = True
    def __init__(self, name):  #define class inintially. Means what parameters will be for class
        self.name = name  #attributes
    
    def run(self):  #create method
        print('run')
        
    @classmethod
    def adding_things(cls,n1,n2):          # 'cls' here is referred to class - same as 'self' in function definition
        return n1 + n2
        
player3 = PlayerBall('vova')
player3.adding_things(2,3)
```

**We can use classmethod without instantiating the class**

```python
PlayerBall.adding_things(2,3)
```

We can use classmethod with class itself:

```python
class PlayerB:
    # Class object attribute
    membership = True
    def __init__(self, name, age):  #define class inintially. Means what parameters will be for class
        self.name = name  #attributes
        self.age = age
    
    def run(self):  #create method
        print('run')
        
    @classmethod
    def adding_things(cls, n1, n2):          # 'cls' here is referred to class - same as 'self' in function definition
        return cls('teddy', n1 + n2)
    
    @staticmethod
    def add_things(num1, num2):
        return num1 + num2
        
player5 = PlayerB.adding_things(2,3)
print(player5.name, player5.age)
```

## Inheritance

It allows a new object to take properites of existion ones

Let's call it to create *subclass*

*subclass* sometimes may be refered to as *child class* or *derived class*

```python
# we have user that can be wizard or archer and so on

class User:
    def sign_in(self):
        print('logged in')
    def attach(self):
        print('do nothing')

class Exemple(User):
    pass
        
class Wizard(User):     # in parentesis we indicate parental class
    def __init__(self, name, power):
        self.name = name
        self.power = power
        
    def attack(self):
        print(f'Attacking with power of {self.power}')

class Archer(User):
    def __init__(self, name, num_arrows):
        self.name = name
        self.num_arrows = num_arrows

    def attack(self):
        print(f'Attacking with arrows. Arrows left: {self.num_arrows}')
```

## Polymorphism
 
 Many forms
 
This idea of polymorphism refers to the way in which object classes can **share** the **same method name** but

those method names can **act differently** based on what object calls the now.

```python
marlin.attack()
robin.attack()
```

Despite the shared name of method - it differ in functionality

Let's create function to do same action

```python
def user_attack(char):
    char.attack()
```

```python
user_attack(marlin)
user_attack(robin)
```

Despite we calling the same function. It gives us different output

Or we can do it inside *for loop*:

```python
for char in[marlin, robin]:
    char.attack()
# we are calling the same method but have different output
```

Note that User class have own *attack* function but it just the default and as long as we define other with same name *attack* in subcalss, the subclasses function will be executed

See one example of how all OOP works

```python
class Pets():
    animals = []
    def __init__(self, animals):
        self.animals = animals

    def walk(self):
        for animal in self.animals:
            print(animal.walk())

class Cat():
    is_lazy = True

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def walk(self):
        return f'{self.name} is just walking around'

class Simon(Cat):
    def sing(self, sounds):
        return f'{sounds}'

class Sally(Cat):
    def sing(self, sounds):
        return f'{sounds}'

#1 Add nother Cat
class Suzy(Cat):
    def sing(self, sounds):
        return f'{sounds}'

#2 Create a list of all of the pets (create 3 cat instances from the above)
my_cats = [Simon('Simon', 4), Sally('Sally', 21), Suzy('Suzy', 1)]

#3 Instantiate the Pet class with all your cats
my_pets = Pets(my_cats)

#4 Output all of the cats singing using the my_pets instance
my_pets.walk()
```

## super() 

To enable parameters from parent class

```python
class Guser():
    def __init__(self, email):
        self.email = email
    
    def sign_in(self):
        print('logged in')
        
class Wizzard(Guser):
    def __init__(self, name, power, email):
        Guser.__init__(self, email)              #we use parent class after def to indicate we need parameter from parent class
        self.name = name
        self.power = power
        
    def attack(self):
        print(f'Attacking with power of {self.power}')

class Archerr(Guser):
    def __init__(self, name, num_arrows, email):
        super().__init__(email)                   # or we can use super() to have same functionality
        self.name = name
        self.num_arrows = num_arrows

    def attack(self):
        print(f'Attacking with arrows. Arrows left: {self.num_arrows}')
```

```python
wizard1 = Wizzard('marlin', 60, 'ddf@dot.com')
```

```python
wizard1.email
```

```python
robinhood = Archerr('robin', 100, 'robin@hood.com')
```

```python
robinhood.email
```

## Multiple inheritance

Turns out we can add as many as we wanr parent classes when creating new one

```python
class Stupid:
    def __init__(self, name, num):
        self.name = name
        self.num = num
        
    def scream(self):
        print('AUUUUU')
        
class Second:
    def __init__(self, name, last_name):
        self.name = name
        self.last_name = last_name
        
    def roar(self):
        print("ROARRRR")
        
class Combiner(Stupid, Second):     # Multiple inheritance
    def __init__(self,name, last_name, num):    # name and num is taken from parent class Stupid
        Second.__init__(self, name, last_name)   # But we need to add last_name from parent class Second

mult_inher = Combiner('Beast', 'Beastovich', 'II')

mult_inher.name
```

```python
mult_inher.roar()
```

```python
mult_inher.scream()
```

```python
mult_inher.last_name
```

## Method Resolution Order (MRO)

```python
class A:
    num = 10
    
class B(A):
    pass

class C(A):
    num = 1
    
class D(B,C):
    pass
```
