# Testing in Python

There is special midule for testing in python called **unittest**

To test python program we actually need to make one more testing program( using unittest module.

Let's assume that we have our main program with one function:

```python
# function below is saved as separate file in same directory as mainfunc.py
'''
def addfunc(num):
    return num + 5
'''
```

Now lets import testing module and our program and 

And write program to test out py prog

Its written as Class

```python
# create another file called testing.py

import unittest
import mainfunc

# create class to test and couple methods for test
class TestMain(unittest.TestCase):
    # create method for test
    def test_our_func(self):
        # declare some test parameter
        test_param = 10
        # call our function with test parameter
        result = mainfunc.addfunc(test_param)
        # here is the test itself:
        # method below i just check that 2 outputs are equal:
        self.assertEqual(result, 15)
        
# to run test we use just following:
unittest.main()
```

Now we can run our test doing following in terminal just by running our *testing.py* file

Now lets upgrade our mainfunc.py and our testing.py files

First let amend some common mistake if enterd string and not int.

```python
# mainfunc.py
def addfunc(num):
    return int(num) + 5
```

If some gibberish is enterd instead of a number the *ValueError* arise

Now lets wrap everuthing in try block

```python
# mainfunc.py
def addfunc(num):
    try:
        return int(num) + 5
    except ValueError as err:
        return err
```

And modify our testing file:

```python
# testing.py file

import unittest
import mainfunc

# create class to test and couple methods for test
class TestMain(unittest.TestCase):
    # create method for test
    def test_our_func(self):
        # declare some test parameter
        test_param = 10
        # call our function with test parameter
        result = mainfunc.addfunc(test_param)
        # here is the test itself:
        # method below i just check that 2 outputs are equal:
        self.assertEqual(result, 15)
    #create 2nd test
    def test_our_func2(self):
        test_param = 'sm_gibberish'
        result = mainfunc.addfunc(test_param)
        # and now we want to test if output will be a ValueError:
        self.assertTrue(isinstance(result, ValueError))
        # Note that we use here isinstance because the output is actually an instance of ValueError class
        
# to run test we use just following:
if __name__ == '__main__':
    unittest.main()
```
