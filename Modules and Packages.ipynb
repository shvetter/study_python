We can create some classes and functions and save it sapartely in .py file. 

Then we can *import* it and use as modules.

```python
# Define some function and save it as file module_example.py
def devid(num1, num2):
    return num1/num2
```

Now we have our own module module_example

```python
import module_example

module_example.devide(1,3)
```

If check what is module_example is:

```python
print(module_example)
```

NOTE that we have generated some folder named *$__pycache__$* in root folder

This folder is used always when we impement *import* as it compiles module and can use it in more efficient way

This is **caching** it's now compiled and now need to do it again

As well as modules, we can create **packages** in Python.

Package is just a folder with different modules.

Let's create one:

```python
# create folder named pack_ex and put her module cart

# module cart will contain following func:
'''
cart = []
def buy(item):
    global cart
    cart.append(item)
    return cart
'''
```

Now we can import module from package in simple way:

```python
from pack_ex.cart import buy
```

```python
buy('snickers')
```

```python
buy('t-shirt')
```

Note that when we crate folder that is going to be our package we need to "explain" it ot Python by putting there **empty**
file called: $__init.py__$

We can also import packages inside of a package with simple . notation

EG we have package named pac1 in and pac2 inside of it, we can access pac2 with following:

import pac1.pac2

## __main__ and __name__

Name of program that is currently opened is "__main__"


```python
print(__name__)
```

Thus it is a good way to start only part of code that is part of following if statement:

```python
if __name__ == '__main__':
    # do something
    # or call some function inside of your code
    print(devid(2,3))
```

So in general - above if statement is usefull just not make a mess while importing. We want to run only code that is inside of main file (inside this statement)

# Build-in modules

https://docs.python.org/3/py-modindex.html

Python have huge build in modules (so called standard library)



Let just see some of buit-in module

**random**

```python
import random

any_ls = [1,2,3,4,5,6,7]
random.shuffle(any_ls)
```

```python
any_ls
```

We can import modules as whatever naame we like to avoid long names:

```python
import random as r

r.shuffle(any_ls)

any_ls
```

Or we can import functions that we going to use, and no need to indicate module name (Best practice)

```python
from random import shuffle

shuffle(any_ls)

any_ls
```

### pip install

We can install from PyPi just from command line. 

pip install is command used to install packages from PyPi

Note that we need to use pip3 if we have older version of Python

We can run **pip -V** to see the version of pip

Use **pip install --upgrade pip ** to install updates

Use **pip list** to see all installed packages


Every module has **version** in common format line 0.5.3

- last digit is some bug fix
- middle is some update like feature of add some functionality
- first is some big update

Sometimes it happens that you project is dependant on some specific version of library. In this case we need to install exact verstion by:

**pip install module_name==0.5.3**

where 0.5.3 is version that you need

Use **pip uninstall madule_name** to delete module or if need to install older version
